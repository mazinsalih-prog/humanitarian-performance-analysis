```python
import pandas as pd

file_path = "Projects Data set.xlsx"

projects = pd.read_excel(file_path, sheet_name="projects")
indicators = pd.read_excel(file_path, sheet_name="indicators")
beneficiaries = pd.read_excel(file_path, sheet_name="beneficiaries")
activities = pd.read_excel(file_path, sheet_name="activities")
```

```python
# Create Achievement Percentage
indicators["achievement"] = indicators["actual"] / indicators["target"]

# Create Gap
indicators["gap"] = indicators["target"] - indicators["actual"]

# Create Performance Status
def classify_performance(x):
    if x >= 0.9:
        return "On Track"
    elif x >= 0.75:
        return "Needs Attention"
    else:
        return "Critical"

indicators["status"] = indicators["achievement"].apply(classify_performance)

indicators.head()
```

```python
print(indicators)
```

```python
# Merge indicators with project info
merged = indicators.merge(projects, on="project_code", how="left")

# View result
merged.head()
```

```python
# Merge indicators with project info
merged = indicators.merge(projects, on="project_code", how="left")

# Show result
merged.head()
```

```python
# Average performance by sector
sector_performance = merged.groupby("sector")["achievement"].mean().sort_values(ascending=False)

sector_performance
```

```python
# Average performance by state
state_performance = merged.groupby("state")["achievement"].mean().sort_values(ascending=False)

state_performance
```

```python
# Count performance status
status_summary = merged["status"].value_counts()

status_summary
```

```python
# Status by sector
status_by_sector = merged.groupby(["sector", "status"]).size().unstack()

status_by_sector
```

```python
import matplotlib.pyplot as plt

sector_performance.plot(kind="bar")
plt.title("Average Performance by Sector")
plt.ylabel("Achievement")
plt.xticks(rotation=45)
plt.show()
```
